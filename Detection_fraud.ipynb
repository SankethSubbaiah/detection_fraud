```python
# Importing Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the dataset to a Pandas Dataframe
credit_data = pd.read_csv('/content/creditcard.csv',error_bad_lines=False)
```

```python
#Print first 5 rows
credit_data.head()
```

```python
#Print last 5 rows
credit_data.tail()
```

```python
#Dataset Info
credit_data.info()
```

```python
#Check for missing values
credit_data.isnull().sum()
```

```python
#Distribution of Valid Transaction and Fraud Transaction
credit_data['Class'].value_counts()
```

**Data Set is Highly unbalanced
0 - Valid Transaction
1 - Fraud Transaction**

```python
valid = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(valid.shape)
print(fraud.shape)
```

```python
valid.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Comparing Values for both Transcation
credit_data.groupby('Class').mean()
```

```python
#Under Sampling
#Build a sample dataset containing similar distribution of Valid transactions and Fraud Transactions

#Number of Fraudulent Transactions --> 352

valid_sample = valid.sample(n=352)
```

```python
new_dataset = pd.concat([valid_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into Features and Targets
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split data into Training and Test Data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training Logistic Regression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Model Evaluation and Accuracy Score
X_train_Prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_Prediction,Y_train)
```

```python
print("Accuracy on Training Data = ",training_data_accuracy)
```

```python
#Accuracy on test data
X_test_Prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_Prediction,Y_test)
```

```python
print("Accuracy on Testing Data = ",testing_data_accuracy)
```
